```python
def cropped_image_2_eyes(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray,1.1,4)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h,x:x+w]
        roi_color = img[y:y+h,x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes)>=2:
            return roi_color
```

```python
crop_image = cropped_image_2_eyes(os.path.join('dataset','sachin_tendulkar','1188471-ftnghlqwyaef3ss.jpg'))
plt.imshow(crop_image)
```

```python
path_data = './dataset/'
path_to_cr_data = './dataset/cropped/'
```

```python
import os 
img_dirs = []
for entry in os.scandir(path_data):
    if entry.is_dir():
        img_dirs.append(entry.path)
```

```python
img_dirs
```

```python
import shutil 
if os.path.exists(path_to_cr_data):
    shutil.rmtree(path_to_cr_data)
os.mkdir(path_to_cr_data)
```

```python
pic = cropped_image_2_eyes('./dataset/cropped/ben_stokes/ben_stokes58.png')
if pic is not None:
    plt.imshow(pic)
```

```python
extensions = ['.jpg','.png','.jpeg','.webp']
```

```python
import pathlib

cropped_image_dirs = []
celebrity_file_names_dict = {}

for img_dir in img_dirs:
    count = 1
    celebrity_name = img_dir.split('/')[-1]
    celebrity_file_names_dict[celebrity_name] = []
    for entry in os.scandir(img_dir):
        if not os.path.isdir(entry.path) and pathlib.Path(entry.path).suffix in extensions :
            roi_color = cropped_image_2_eyes(entry.path)
            if roi_color is not None:
                cropped_folder = path_to_cr_data + celebrity_name
                if not os.path.exists(cropped_folder):
                    os.makedirs(cropped_folder)
                    cropped_image_dirs.append(cropped_folder)
                    print("Generating cropped images in folder: ",cropped_folder)
                cropped_file_name = celebrity_name + str(count) + ".png"
                cropped_file_path = cropped_folder + "/" + cropped_file_name
                cv2.imwrite(cropped_file_path, roi_color)
                celebrity_file_names_dict[celebrity_name].append(cropped_file_path)
                count += 1
    
```

```python
import numpy as np
import pywt
import cv2    

def w2d(img, mode='haar', level=1):
    imArray = img
    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255
    coeffs=pywt.wavedec2(imArray, mode, level=level)

    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python
harr_img = w2d(pic,'db1',5)
plt.imshow(harr_img,cmap='gray')
```

```python
del celebrity_file_names_dict['cropped']
```

```python
celebrity_file_names_dict.keys()
```

```python
class_dict = {}
count = 0
for celeb_cls in celebrity_file_names_dict.keys():
    class_dict[celeb_cls] = count
    count+=1
```

```python
class_dict
```

```python
X=[]
y=[]

for celeb_name,files in celebrity_file_names_dict.items():
    for file in files:
        img = cv2.imread(file)
        if img is not None:
            scaled_img = cv2.resize(img,(32,32))
            harr = w2d(img,'db1',5)
            scaled_harr = cv2.resize(harr,(32,32))
            stacked = np.vstack((scaled_img.reshape(32*32*3,1),scaled_harr.reshape(32*32*1,1)))
            X.append(stacked)
            y.append(class_dict[celeb_name])
```

```python
X = np.array(X).reshape(len(X),4096).astype(float)
```

```python
import pandas as pd
df = pd.DataFrame(X)
df_y = pd.Series(y)
```

```python
df['target']=df_y.values
```

```python
df.to_csv('dataset.csv')
```
